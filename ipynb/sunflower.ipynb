```python
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sunflower(N=100,axis=0,mode="equal_size"):
    if mode=="grow_size":
        s=[20*i for i in range(1,N+1)]
        def radius(i):
            return i
    else:
        s=[100 for i in range(1,N+1)]
        def radius(i):
            return np.sqrt(i)
    x = []
    y = []
    phi=(np.sqrt(5)+1)/2
    thetas=[i*np.pi*2*phi for i in range(N)]
    for i in range(N):
        theta = thetas[i]
        r =  radius(i)
        xi = r * np.cos(theta)
        yi = r * np.sin(theta)
        x.append(xi)
        y.append(yi)

    m = max(max([abs(i)for i in x]),max([abs(i) for i in y]))
    plt.figure(figsize=(8, 8))
    if axis==1:
        plt.plot([-1.1*m,1.1*m],[0,0],'k',linewidth=0.5)  # x axis
        plt.plot([0,0],[-1.1*m,1.1*m],'k',linewidth=0.5)  # y axis
    plt.scatter(x,y,s=s,c='gold',marker='o')
    plt.show()
    #plt.savefig("%d.jpg"%N)
```

```python
sunflower(10,axis=1,mode="grow_size")
```

```python
sunflower(20,axis=1,mode="grow_size")
```

```python
sunflower(100,mode="grow_size")
```
